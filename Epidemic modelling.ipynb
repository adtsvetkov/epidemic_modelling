Обучить модель с экспоненциальным ростом числа заболевших:

$$y \sim \exp(linreg) \Leftrightarrow \ln{y} \sim linreg$$

```python
import numpy as np

class MyLinReg_exp:
    def __init__(self, X, Y, exp=True, free_term=True):
        self.free_term = free_term
        if exp:
            self.Y = np.log(np.array(Y).T)
        else:
            self.Y = np.array(Y).T
        if self.free_term:
            self.X = np.c_[np.array(X), np.ones(len(X))]
        else:
            self.X = np.array(X).reshape(-1, 1)
        self.a = None
    def fit(self):
        self.a = np.linalg.inv(self.X.T.dot(self.X)).dot(self.X.T).dot(self.Y)
    def predict(self, X):
        if self.free_term:
            X = np.c_[np.array(X), np.ones(len(X))]
        return np.sum(X * self.a, axis=1)
    def get_a(self):
        return self.a
```

```python
# сначала запредиктим total_cases

lr = MyLinReg_exp(data_train['day_number'].values,
              data_train['total_cases'].values)
lr.fit()
lr1_pred = lr.predict(data_test['day_number'].values)
```

Найти апостериорное распределение параметров модели для достаточно широкого априорного распределения. Требующееся для этого значение дисперсии шума в данных оценить, исходя из построенной максимальной апостериорной модели.

Введем нормальное распределение для априорного распределения:
$$p(w) = N(\mu_0, \Sigma_0)$$
$$\Sigma_N = \left(\Sigma_0^{-1} + \frac{X^TX}{\sigma^2}\right)^{-1}$$
$$\overline{\mu}_N = \Sigma_N\left(\Sigma_0^{-1}\overline{\mu}_0 + \frac{X^T\overline{y}}{\sigma^2}\right)$$

Как делали на лекции, возьмём $\mu = (0, 0), \sigma = M\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}$ (пусть $M=250$)

Значение шума посчитаем как стандартное отклонение реальных значений и тех, которые получаются при МНК-оценке коэффициентов линейной регрессии:
$$\sigma = std(Y - X((X^TX)^{-1}X^TY)$$

```python
# посчитаем апостериорное распределение

def bayesian_update(mu, sigma, x, y, sigma_noise=None):
    y = np.array(y)
    if not sigma_noise:
        # добавим к x столбец с 1, чтобы иметь свободный член в предсказаниях
        x = np.c_[np.array(x), np.ones(len(x))]
        sigma_noise = np.std(y - np.sum(x * (np.linalg.inv(x.T.dot(x)).dot(x.T).dot(y)), axis=1))
    sigma_n = np.linalg.inv(np.linalg.inv(sigma) + (1 / (sigma_noise ** 2)) * np.dot(np.transpose(x), x))
    mu_n = np.dot(sigma_n, np.dot(np.linalg.inv(sigma), np.transpose(mu)) + (1 / (sigma_noise ** 2)) * np.dot(np.transpose(x), y))
    
    return mu_n, sigma_n
```

```python
main_mu = np.array([0, 0])
main_sigma = np.array([[250, 0], [0, 250]])

# new_cases тренировочные данные
mu_estimation_nc, sigma_estimation_nc = bayesian_update(main_mu, main_sigma, 
                                                 data_train['day_number'].values,
                                                 np.log(new_cases_train))
# total_cases тренировочные данные
mu_estimation_tc, sigma_estimation_tc = bayesian_update(main_mu, main_sigma, 
                                                 data_train['day_number'].values,
                                                 np.log(data_train['total_cases']))
```

Сколько, исходя из этих сэмплов, предсказывается случаев коронавируса в России к 1 мая? к 1 июня? к 1 сентября?

```python
import math

def get_pred_by_date(df, date, column, mu, sigma, sample=1, info=False):
    rows_needed = df.loc[df['date'] == date]
    x, y = rows_needed['day_number'], rows_needed[column].item()
    x = np.array([x, 1], dtype=np.float64)
    my_w = np.random.multivariate_normal(mu, sigma, sample)
    predictions = [np.exp(x.dot(w)) for w in my_w]
    
    if info:
        pred = math.ceil(np.mean(predictions))
        line = [date, int(x[0]), column, int(y), pred,
               abs(y - pred)]
        return line
    
    return math.ceil(np.mean(predictions))
```

```python
col_names = ['Дата', 'Номер дня', 'Колонка', 'Реальное значение', 'Предсказанное значение', 'Ошибка']
results = pd.DataFrame(columns=col_names)

for d in ['2020-05-01', '2020-06-01', '2020-09-01']:
    line1 = get_pred_by_date(data_Russia, d,
                            'new_cases', mu_estimation_nc, sigma_estimation_nc, info=True)
    results = pd.concat([results, pd.DataFrame([line1], columns=col_names)])
    
    line2 = get_pred_by_date(data_Russia, d,
                            'total_cases', mu_estimation_tc, sigma_estimation_tc, info=True)
    results = pd.concat([results, pd.DataFrame([line2], columns=col_names)])

results = results.reset_index(drop=True)
results
```

## Шаг 3. Сигмоидальная функция

Проделаем то же самое, что и в прошлом шаге, но уже с сигмоидальной функцией, а не с экспонентой.

Предложить способ обучать параметры такой сигмоидальной функции при помощи линейной регрессии.

$$Ф(x) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{x}{e^{\frac{-t^2}{2}}dt}$$

Т.к. мы допускаем, что константа перед интегралом может быть произвольной и ее можно внести в экспоненту, а вместо t есть любая квадратичная функция от t, то примем следующее утверждение:

$$y = \int_{-\infty}^{x}{e^{t}}$$
$$t = a_2x^2 + a_1x + a_0$$

Т.к. мы хотим справа получить логрег, сделаем так:
$$\ln{\nabla y} = a_2x^2 + a_1x + a_0$$

Теперь нужно понять, где взять $\nabla y$. По определению производная:
$$\lim_{\Delta t \rightarrow 0}{\frac{y(t+\Delta t) - y(t)}{\Delta t}}$$

То есть предел по приращению времени. Время у нас измеряется в сутках, тогда получим для $\Delta t = 1$:

$$\lim{y_{k+1} - y_k} = \lim_{\Delta t \rightarrow 0}{y_{new}}$$

То есть производная всех случаев есть приращение новых случаев.
$$\nabla y_{total} \sim y_{new}$$

Обучить эти параметры на датасете случаев коронавируса в России.

```python
X_ = np.array(data_Russia['day_number'].values)

X = np.c_[X_, X_ ** 2]

X_train, X_test = X[:train_size], X[train_size:]
```

```python
import numpy as np

class MyLinReg_sigm:
    def __init__(self, X, Y, exp=True, free_term=True):
        self.free_term = free_term
        if exp:
            self.Y = np.log(np.array(Y).T)
        else:
            self.Y = np.array(Y).T
        if self.free_term:
            self.X = np.c_[np.array(X), np.ones(X.shape[0])]
        else:
            self.X = np.array(X).reshape(-1, 1)
        self.a = None
    def fit(self):
        self.a = np.linalg.inv(self.X.T.dot(self.X)).dot(self.X.T).dot(self.Y)
    def predict(self, X):
        if self.free_term:
            X = np.c_[np.array(X), np.ones(X.shape[0])]
        return np.sum(X * self.a, axis=1)
    def get_a(self):
        return self.a
```

```python
lr = MyLinReg_sigm(X_train, new_cases_train)
lr.fit()
lr1_pred = lr.predict(X_test)
```

Провести такой же анализ для других стран. Насколько разные параметры получаются? Можно ли разделить страны на кластеры (хотя бы чисто визуально) в зависимости от этих параметров?

Для решения этой задачи посчитаем параметры линрега экспоненты и сигмоиды, а затем попытаемся кластеризовать их.

```python
# получаем нужную информацию
train_size = 50

results = []

for country in data.location.unique():
    local_res = [country]
    dataset = data.loc[data['location'] == country]
    local_res.append(dataset['continent'].values[0])
    dataset = dataset[dataset['total_cases'] > 2.0]
    dataset = dataset[dataset['total_cases'].notna()]
    dataset['day_number'] = range(1, dataset.shape[0] + 1)
    
    dataset_train = dataset.head(train_size)
    X_ = np.array(dataset_train['day_number'].values)
    X_train = np.c_[X_, X_ ** 2]
    
    local_res.append(dataset_train)
    
    new_cases = dataset_train['new_cases'].replace(0, 1)
    new_cases = new_cases.replace(np.nan, 1)
    
    if dataset_train.shape[0] < train_size or np.all(new_cases == 1.0):
        continue
    
    # экспонента
    lr = MyLinReg_exp(dataset_train['day_number'].values,
              dataset_train['total_cases'].values)
    lr.fit()
    lr1_pred = lr.predict(dataset_train['day_number'].values)
    
    local_res.extend(lr.get_a())
    
    main_mu = np.array([0, 0])
    main_sigma = np.array([[250, 0], [0, 250]])
    
    mu_estim_exp, sigma_estim_exp = bayesian_update(main_mu,
                                                    main_sigma,
                                                    dataset_train['day_number'].values,
                                                    np.log(dataset_train['total_cases']))
    
    local_res.extend([mu_estim_exp, sigma_estim_exp])
    
    # сигмоида
    lr = MyLinReg_sigm(X_train, new_cases)
    lr.fit()
    lr1_pred = lr.predict(X_train)
    
    local_res.extend(lr.get_a())
    
    main_mu = np.array([0, 0, 0])
    main_sigma = np.array([[1000, 0, 0], [0, 1000, 0], [0, 0, 1000]])
    
    mu_estim_sigm, sigma_estim_sigm = bayesian_update(main_mu, main_sigma,
                                                     X_train,
                                                     np.log(new_cases))
    
    local_res.extend([mu_estim_sigm, sigma_estim_sigm])
    results.append(local_res)

plt.show()
```

```python
plotting_data = pd.DataFrame(results, 
                    columns=['Страна', 'Континент', 'Train данные', 'Exp_a1', 'Exp_a2',
                            'mu_exp', 'sigma_exp', 'Sigm_a1', 'Sigm_a2', 'Sigm_a3',
                            'mu_sigm', 'sigma_sigm'])
```
